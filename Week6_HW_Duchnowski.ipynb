### Assignments
* Create a visualization of K-means: locations, labels, and cluster centers.

```python
#Data Transformation
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn import preprocessing
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
#Get data
X  = load_iris().data

#Scale data
X = preprocessing.scale(X)

#Create model
kmeans_cluster = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
#Predict the iris species (or "label")
label = kmeans_cluster.fit_predict(X)

#Loop thru unique labels and plot: 
for i in np.unique(label):
    plt.scatter(X[label == i , 0], X[label == i , 1], label = i, s = 120, alpha=0.4 )
    
#Plot the cluster centers and overlay: 
plt.scatter(kmeans_cluster.cluster_centers_[:,0] , kmeans_cluster.cluster_centers_[:,1], s = 120, color = 'black', alpha=1.0)

plt.legend()
plt.show()
```
